```python
def gencube(n):
    for i in range(n):
        return n**3
```

```python
gencube(4)
```

```python
def gencube1(n):
    l = []
    for i in range(n):
        l.append(i**3)
    return l
```

```python
gencube1(5)
```

```python
gencube1(1000)
```

```python
# as gencube1 fun is taking time bcoz we're asking result to be returned in list. 
#Hence we need to use range fun,as it doesn't store data and give result immediately.
```

```python
def gencube2(n):
    for i in range(n):
        yield i**3
```

```python
gencube2(100)
```

```python
# As gencube fun is converted as generator,to get output which is generated in above line
for i in gencube2(10):
    print(i)
```

```python
"""In above line print fun is used to get output, 
if print fun used inside gencube def then it gives NoneType not in for loop after gencube def """
```

```python
# Generation of Fibnacci sequence (where number= Sum of previous 2 no's). Eg:0 1 1 2 3 5 8 13 21 44,etc.,
def fib(n):
    a=1
    b=1
    for i in range(n):
        yield a,i
        a,b = b,a+b
```

```python
for i in fib(10):
    print(i)
```

```python
# Another way to write fibnacci series without using Yield
def fib1(n):
    a=1
    b=1
    output=[]
    for i in range(n):
        output.append(a)
        a,b = b,a+b
    return output
```

```python
fib1(10)
```
