```python
t = (1,"ffgf",5,8,65,741,54.25)
for i in range(len(t)-1,-1,-1):
    print(i,t[i])
```

```python
# range(len(t)) - Gives you no.of elements present in list/tuple, starting from index 1 but not zero.
# Hence to match with index's of list/tuple, it has to be written as range(len(t)-1)
```

```python
t = (1,"ffgf",5,8,65,741,54.25)
for i in range(len(t)-1,,-1):
    print(i,t[i])
```

```python
# So range always require lower bound and upper bound to be given
```
